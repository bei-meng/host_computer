```python
import matplotlib.pyplot as plt
import numpy as np
import json 
```

```python
path = "./result/read_data/row_read_10.json"
with open(path, 'r') as f:
    data = json.load(f)

cross_cond = np.zeros((256, 256))
for row,row_cond in data.items():
    print(row)
    tmp = []
    for k in range(8):
        # 每次选两路TIA的值进行合并,如1:2:31,2:2:32合并为1:1:32
        for i in range(16):
            tmp.append(row_cond[str(k)][i])#第1列
            tmp.append(row_cond[str(k+8)][i])#第2列
    flag = 0
    for i in range(256):
        if tmp[i]<0 or tmp[i]>80:
            flag = 1
            continue
        cross_cond[int(row),i]=tmp[i]
    if flag:
        print("有坏点:",row)
# 绘制矩阵
plt.figure(figsize=(24,20))
plt.imshow(cross_cond)  # 你可以选择其他的colormap, 例如'viridis', 'plasma', 'inferno'等
cbar  = plt.colorbar()  # 显示色标
cbar.set_label("us")
plt.title('256x256 cond.(us)')  # 设置标题
plt.xlabel('col')  # 设置X轴标签
plt.ylabel('row')  # 设置Y轴标签
plt.show()  # 显示图像
```
